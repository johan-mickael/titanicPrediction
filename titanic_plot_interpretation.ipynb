```python
# Bar Plot of Survival Counts by Gender
survival_counts = data.groupby('sex')['survived'].value_counts().unstack()
survival_counts.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Survival Counts by Gender')
plt.legend(['Did Not Survive', 'Survived'])
plt.show()
```

```python
# Pie Chart of Survival Rate by Passenger Class
survival_rate = data.groupby('pclass')['survived'].mean()
plt.pie(survival_rate, labels=['1st Class', '2nd Class', '3rd Class'], autopct='%1.1f%%')
plt.title('Survival Rate by Passenger Class')
plt.show()
```
